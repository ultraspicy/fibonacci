```python
# Run once
import os
import requests
from pathlib import Path
from urllib.parse import urlparse

def download_videos(categories_list, output_folder="downloaded_videos"):
    # Create output folder if it doesn't exist
    Path(output_folder).mkdir(parents=True, exist_ok=True)
    
    total_videos = 0
    downloaded = 0
    failed = 0
    
    # Iterate through categories
    for category in categories_list:
        category_name = category.get("name", "Unknown")
        print(f"\nProcessing category: {category_name}")
        
        # Iterate through videos in each category
        for video in category.get("videos", []):
            title = video.get("title", "Unknown")
            sources = video.get("sources", [])
            
            total_videos += 1
            
            # Download from each source (usually just one)
            for source_url in sources:
                try:
                    # Extract filename from URL
                    parsed_url = urlparse(source_url)
                    filename = os.path.basename(parsed_url.path)
                    
                    # Create full path for the file
                    filepath = os.path.join(output_folder, filename)
                    
                    # Skip if file already exists
                    if os.path.exists(filepath):
                        print(f"  ✓ '{title}' already exists, skipping...")
                        downloaded += 1
                        continue
                    
                    print(f"  Downloading '{title}'...")
                    
                    # Download the video with streaming to handle large files
                    response = requests.get(source_url, stream=True, timeout=30)
                    response.raise_for_status()
                    
                    # Get file size if available
                    total_size = int(response.headers.get('content-length', 0))
                    
                    # Write to file in chunks
                    with open(filepath, 'wb') as f:
                        if total_size == 0:
                            f.write(response.content)
                        else:
                            downloaded_size = 0
                            for chunk in response.iter_content(chunk_size=8192):
                                if chunk:
                                    f.write(chunk)
                                    downloaded_size += len(chunk)
                                    # Show progress
                                    percent = (downloaded_size / total_size) * 100
                                    print(f"    Progress: {percent:.1f}%", end='\r')
                    
                    print(f"  ✓ Successfully downloaded '{title}'")
                    downloaded += 1
                    
                except requests.exceptions.RequestException as e:
                    print(f"  ✗ Failed to download '{title}': {e}")
                    failed += 1
                except Exception as e:
                    print(f"  ✗ Error downloading '{title}': {e}")
                    failed += 1
    
    # Print summary
    print(f"\n{'='*50}")
    print(f"Download Summary:")
    print(f"  Total videos: {total_videos}")
    print(f"  Successfully downloaded: {downloaded}")
    print(f"  Failed: {failed}")
    print(f"  Output folder: {os.path.abspath(output_folder)}")
    print(f"{'='*50}")



download_videos(test_videos)
```

```python
import cv2
import numpy as np
import os

def compute_channel_change_fractions(video_path):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise RuntimeError(f"Could not open video: {video_path}")

    changed_fractions = []

    ret, prev_frame = cap.read()
    if not ret:
        cap.release()
        raise RuntimeError(f"Could not read first frame of: {video_path}")

    prev_frame = prev_frame.astype(np.uint8)

    while True:
        ret, curr_frame = cap.read()
        if not ret:
            break

        curr_frame = curr_frame.astype(np.uint8)

        # Compute per-channel change fraction
        for ch in range(3):
            prev_ch = prev_frame[:, :, ch]
            curr_ch = curr_frame[:, :, ch]

            diff_ch = prev_ch != curr_ch
            fraction_ch = diff_ch.mean()

            changed_fractions.append(fraction_ch)

        prev_frame = curr_frame

    cap.release()
    return changed_fractions

VIDEO_DIR = "downloaded_videos"

optimistic_estimates = []
realistic_estimates = []

for fname in sorted(os.listdir(VIDEO_DIR)):
    if not fname.lower().endswith((".mp4", ".mov", ".avi", ".mkv")):
        continue

    full_path = os.path.join(VIDEO_DIR, fname)
    print(f"Processing: {fname}")

    fractions = compute_channel_change_fractions(full_path)
    print(f"  Computed {len(fractions)} channel-changes ({len(fractions)//3} frame transitions)")

    work_counter = 0
    for val in fractions:
        if val <= 0.05:
            work_counter += (0.05)*2.5
        elif val <= 0.1:
            work_counter += (0.1)*2.5
        elif val <= 0.2:
            work_counter += (0.2)*2.5
        else:
            work_counter += 1
    print(f"Reasonable estimated work for file name: {fname} is {work_counter/len(fractions)}")
    realistic_estimates.append(work_counter / len(fractions))
    
    work_counter = 0
    for val in fractions:
        if val <= 0.4:
            work_counter += val*2.5
        else:
            work_counter += 1
    print(f"Optimistic estimated work for file name: {fname} is {work_counter/len(fractions)}")
    optimistic_estimates.append(work_counter / len(fractions))
```

```python
def print_stats(values):
    if not values:
        print("List is empty.")
        return

    minimum = min(values)
    maximum = max(values)
    average = sum(values) / len(values)

    print(f"Min: {minimum}")
    print(f"Max: {maximum}")
    print(f"Avg: {average}")
    
print("Realistic estimates stats:")
print_stats(realistic_estimates)

print("Optimistic estimates stats:")
print_stats(optimistic_estimates)
```
