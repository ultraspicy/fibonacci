```python
import sys
import numpy as np
from PIL import Image

def decompose_image(file_path, save_path):
    # Open the image file
    img = Image.open(file_path)
    
    # Convert the image to a NumPy array
    img_array = np.array(img)
    
    # Check if the image is RGB
    if len(img_array.shape) != 3 or img_array.shape[2] != 3:
        raise ValueError("The image must be in RGB format")
    
    # Decompose into R, G, B channels
    r_channel = img_array[:, :, 0]
    g_channel = img_array[:, :, 1]
    b_channel = img_array[:, :, 2]
    
    for channel, name in zip([r_channel, g_channel, b_channel], ['R', 'G', 'B']):
        output_file = f'{save_path}_{name}.txt'
        np.savetxt(output_file, channel, fmt='%d', delimiter=' ')
        print(f"Saved {name} channel to {output_file}")
    
    return r_channel, g_channel, b_channel

for i in sorted(os.listdir("../demo/frames")):
    if i.endswith(".png"):
        r,g,b = decompose_image(os.path.join("../demo/frames", i), os.path.join("../demo/decomposed_frames", i[:-4]))
```

```python
import json
import time
import os
import warnings
import sys
import numpy as np
from PIL import Image

# Directory containing your decomposed text files
decomposed_dir = "../demo/decomposed_frames"

# Collect all base frame names (without _R.txt, etc.)
all_frames = sorted(set(f[:-6] for f in os.listdir(decomposed_dir) if f.endswith("_R.txt")))

# Limit to first 1000 frames
frame_subset = all_frames[0:1500]

r_channel = []
g_channel = []
b_channel = []

for base in frame_subset:
    r_path = os.path.join(decomposed_dir, f"{base}_R.txt")
    g_path = os.path.join(decomposed_dir, f"{base}_G.txt")
    b_path = os.path.join(decomposed_dir, f"{base}_B.txt")

    r_channel.append(np.loadtxt(r_path))
    g_channel.append(np.loadtxt(g_path))
    b_channel.append(np.loadtxt(b_path))

print(f"Loaded {len(r_channel)} frames for each channel.")
```

```python
import numpy as np

def fraction_changes(channel_list):
    """Compute fraction of changed pixels between each pair of consecutive frames."""
    fractions = []
    for i in range(1, len(channel_list)):
        frame_prev = channel_list[i - 1]
        frame_curr = channel_list[i]
        
        # Ensure same shape
        if frame_prev.shape != frame_curr.shape:
            raise ValueError(f"Frame size mismatch at index {i}")
        
        # Count changed pixels
        changed = np.sum(frame_prev != frame_curr)
        total = frame_prev.size
        fractions.append(changed / total)
    return np.array(fractions)

# Compute fractions for each color channel
r_changes = fraction_changes(r_channel)
g_changes = fraction_changes(g_channel)
b_changes = fraction_changes(b_channel)

threshold = 0.10  # 20%

r_stable = np.sum(r_changes < threshold)
g_stable = np.sum(g_changes < threshold)
b_stable = np.sum(b_changes < threshold)

print(f"Number of frame transitions with <20% pixel change:")
print(f"R: {r_stable} / {len(r_changes)}")
print(f"G: {g_stable} / {len(g_changes)}")
print(f"B: {b_stable} / {len(b_changes)}")


def average_change_amplitude(channel_list, changes, threshold):
    """Compute mean abs diff among changed pixels for frames with <threshold fraction change."""
    per_pair_avgs = []

    for i in range(1, len(channel_list)):
        if changes[i - 1] < threshold:
            frame_prev = channel_list[i - 1]
            frame_curr = channel_list[i]
            diff = np.abs(frame_curr.astype(float) - frame_prev.astype(float))
            changed_mask = diff > 0
            if np.any(changed_mask):
                avg_diff = diff[changed_mask].mean()
                per_pair_avgs.append(avg_diff)

    return np.mean(per_pair_avgs) if per_pair_avgs else float('nan')

r_avg_change = average_change_amplitude(r_channel, r_changes, threshold)
g_avg_change = average_change_amplitude(g_channel, g_changes, threshold)
b_avg_change = average_change_amplitude(b_channel, b_changes, threshold)


print("Average |pixel| among changed pixels (only for 'close' frames):")
print(f"R: {r_avg_change:.2f}")
print(f"G: {g_avg_change:.2f}")
print(f"B: {b_avg_change:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

threshold = 0.20  # 10%

def collect_differences_for_stable_frames(channel_list, changes, threshold):
    """Collect all |pixel| values for frames with <threshold fraction change."""
    diffs = []

    for i in range(1, len(channel_list)):
        if changes[i - 1] < threshold:
            frame_prev = channel_list[i - 1].astype(float)
            frame_curr = channel_list[i].astype(float)
            diff = np.abs(frame_curr - frame_prev)
            # Only keep changed pixels
            diffs.append(diff[diff > 0])

    if diffs:
        return np.concatenate(diffs)
    else:
        return np.array([])

# Collect diffs for each color channel
r_diffs = collect_differences_for_stable_frames(r_channel, r_changes, threshold)
g_diffs = collect_differences_for_stable_frames(g_channel, g_changes, threshold)
b_diffs = collect_differences_for_stable_frames(b_channel, b_changes, threshold)

print(f"Collected {len(r_diffs):,} red, {len(g_diffs):,} green, {len(b_diffs):,} blue pixel differences.")

# Filter out large differences (>100)
r_diffs_filtered = r_diffs[r_diffs <= 50]
g_diffs_filtered = g_diffs[g_diffs <= 50]
b_diffs_filtered = b_diffs[b_diffs <= 50]

# Plot histograms again
plt.figure(figsize=(10, 6))
plt.hist(r_diffs_filtered, bins=50, alpha=0.6, label='Red channel', density=True)
plt.hist(g_diffs_filtered, bins=50, alpha=0.6, label='Green channel', density=True)
plt.hist(b_diffs_filtered, bins=50, alpha=0.6, label='Blue channel', density=True)

plt.title(f"Histogram of |pixel| for stable frame transitions (<{threshold*100:.0f}% change, values  100)")
plt.xlabel("Absolute pixel difference")
plt.ylabel("Density")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```
